- æœ¬è¯¾ç¨‹ä½¿ç”¨Claude 3 Haikuï¼Œæ¸©åº¦è®¾ç½®ä¸º0ã€‚æˆ‘ä»¬å°†åœ¨è¯¾ç¨‹åé¢è¯¦ç»†è®¨è®ºæ¸©åº¦å‚æ•°ã€‚ç°åœ¨ï¼Œåªéœ€ç†è§£è¿™äº›è®¾ç½®èƒ½äº§ç”Ÿæ›´ç¡®å®šæ€§çš„ç»“æœã€‚æœ¬è¯¾ç¨‹ä¸­çš„æ‰€æœ‰æç¤ºå·¥ç¨‹æŠ€æœ¯ä¹Ÿé€‚ç”¨äºä¹‹å‰çš„ä¼ ç»ŸClaudeæ¨¡å‹ï¼Œå¦‚Claude 2å’ŒClaude Instant 1.2ã€‚

### Anthropic SDKå’ŒMessages API
æˆ‘ä»¬å°†åœ¨æ•´ä¸ªæ•™ç¨‹ä¸­ä½¿ç”¨[Anthropic Python SDK](https://docs.anthropic.com/claude/reference/client-sdks)å’Œ[Messages API](https://docs.anthropic.com/claude/reference/messages_post)ã€‚

ä¸‹é¢æ˜¯æœ¬æ•™ç¨‹ä¸­è¿è¡Œæç¤ºçš„ç¤ºä¾‹ã€‚é¦–å…ˆï¼Œæˆ‘ä»¬åˆ›å»º`get_completion`ï¼Œè¿™æ˜¯ä¸€ä¸ªè¾…åŠ©å‡½æ•°ï¼Œç”¨äºå‘Claudeå‘é€æç¤ºå¹¶è¿”å›Claudeç”Ÿæˆçš„å“åº”ã€‚ç°åœ¨è¿è¡Œè¯¥å•å…ƒæ ¼ã€‚

```python
# å¯¼å…¥Anthropicåº“
import anthropic

# åˆ›å»ºAnthropicå®¢æˆ·ç«¯å®ä¾‹
client = anthropic.Anthropic(api_key=API_KEY)

def get_completion(prompt: str):
    """
    è¾…åŠ©å‡½æ•°ï¼šå‘Claudeå‘é€æç¤ºå¹¶è·å–å›å¤
    
    å‚æ•°:
        prompt (str): å‘é€ç»™Claudeçš„æç¤ºæ–‡æœ¬
    
    è¿”å›:
        str: Claudeçš„å›å¤æ–‡æœ¬
    """
    message = client.messages.create(
        model=MODEL_NAME,           # ä½¿ç”¨çš„æ¨¡å‹åç§°
        max_tokens=2000,           # æœ€å¤§ç”Ÿæˆtokenæ•°
        temperature=0.0,           # æ¸©åº¦å‚æ•°ï¼Œ0è¡¨ç¤ºæ›´ç¡®å®šæ€§çš„è¾“å‡º
        messages=[
          {"role": "user", "content": prompt}  # ç”¨æˆ·è§’è‰²çš„æ¶ˆæ¯
        ]
    )
    return message.content[0].text
```

ç°åœ¨æˆ‘ä»¬å°†ä¸ºClaudeç¼–å†™ä¸€ä¸ªç¤ºä¾‹æç¤ºï¼Œå¹¶é€šè¿‡è¿è¡Œæˆ‘ä»¬çš„`get_completion`è¾…åŠ©å‡½æ•°æ¥æ‰“å°Claudeçš„è¾“å‡ºã€‚è¿è¡Œä¸‹é¢çš„å•å…ƒæ ¼å°†åœ¨å…¶ä¸‹æ–¹æ‰“å°å‡ºClaudeçš„å“åº”ã€‚

è¯·éšæ„ä¿®æ”¹æç¤ºå­—ç¬¦ä¸²ï¼Œä»¥è·å¾—Claudeçš„ä¸åŒå“åº”ã€‚

```python
# æç¤ºæ–‡æœ¬
prompt = "Hello, Claude!"

# è·å–Claudeçš„å“åº”
print(get_completion(prompt))
```

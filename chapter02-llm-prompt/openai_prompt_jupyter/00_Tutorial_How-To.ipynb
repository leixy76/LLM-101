3. è®¾ç½®æ‚¨çš„APIå¯†é’¥å’Œæ¨¡å‹åç§°ã€‚å°†`"your_api_key_here"`æ›¿æ¢ä¸ºæ‚¨å®é™…çš„OpenAI APIå¯†é’¥ã€‚

```python
# è®¾ç½®APIå¯†é’¥å’Œæ¨¡å‹åç§°
API_KEY = "your_api_key_here"  # è¯·æ›¿æ¢ä¸ºæ‚¨çš„å®é™…OpenAI APIå¯†é’¥
MODEL_NAME = "gpt-4o"  # GPT-4oæ¨¡å‹ï¼ˆæ¨èï¼‰æˆ– "deepseek-r1"

# å°†API_KEYå’ŒMODEL_NAMEå˜é‡å­˜å‚¨åœ¨IPythonå­˜å‚¨ä¸­ï¼Œä»¥ä¾¿åœ¨ä¸åŒnotebookä¸­ä½¿ç”¨
%store API_KEY
%store MODEL_NAME
```

## ä½¿ç”¨è¯´æ˜å’Œæç¤º ğŸ’¡

- æœ¬è¯¾ç¨‹ä½¿ç”¨GPT-4oï¼Œæ¸©åº¦è®¾ç½®ä¸º0ã€‚æˆ‘ä»¬å°†åœ¨è¯¾ç¨‹åé¢è¯¦ç»†è®¨è®ºæ¸©åº¦å‚æ•°ã€‚ç°åœ¨ï¼Œåªéœ€ç†è§£è¿™äº›è®¾ç½®èƒ½äº§ç”Ÿæ›´ç¡®å®šæ€§çš„ç»“æœã€‚æœ¬è¯¾ç¨‹ä¸­çš„æ‰€æœ‰æç¤ºå·¥ç¨‹æŠ€æœ¯ä¹Ÿé€‚ç”¨äºå…¶ä»–OpenAIæ¨¡å‹ï¼Œå¦‚GPT-4o-miniï¼Œä»¥åŠDeepSeekçš„deepseek-r1ç­‰æ¨¡å‹ã€‚

- æ‚¨å¯ä»¥ä½¿ç”¨`Shift + Enter`æ‰§è¡Œå•å…ƒæ ¼å¹¶ç§»åŠ¨åˆ°ä¸‹ä¸€ä¸ªã€‚

- å½“æ‚¨åˆ°è¾¾æ•™ç¨‹é¡µé¢åº•éƒ¨æ—¶ï¼Œå¯¼èˆªåˆ°æ–‡ä»¶å¤¹ä¸­çš„ä¸‹ä¸€ä¸ªç¼–å·æ–‡ä»¶ï¼Œæˆ–è€…å¦‚æœæ‚¨å·²å®Œæˆè¯¥ç« èŠ‚æ–‡ä»¶å†…çš„å†…å®¹ï¼Œåˆ™è½¬åˆ°ä¸‹ä¸€ä¸ªç¼–å·æ–‡ä»¶å¤¹ã€‚

### OpenAI SDKå’ŒChat Completions API
æˆ‘ä»¬å°†åœ¨æ•´ä¸ªæ•™ç¨‹ä¸­ä½¿ç”¨[OpenAI Python SDK](https://platform.openai.com/docs/libraries)å’Œ[Chat Completions API](https://platform.openai.com/docs/api-reference/chat)ã€‚

ä¸‹é¢æ˜¯æœ¬æ•™ç¨‹ä¸­è¿è¡Œæç¤ºçš„ç¤ºä¾‹ã€‚é¦–å…ˆï¼Œæˆ‘ä»¬åˆ›å»º`get_completion`ï¼Œè¿™æ˜¯ä¸€ä¸ªè¾…åŠ©å‡½æ•°ï¼Œç”¨äºå‘GPTå‘é€æç¤ºå¹¶è¿”å›GPTç”Ÿæˆçš„å“åº”ã€‚ç°åœ¨è¿è¡Œè¯¥å•å…ƒæ ¼ã€‚

```python
# å¯¼å…¥OpenAIåº“
import openai

# å¦‚æœæ²¡æœ‰è®¾ç½®MODEL_NAMEï¼Œä½¿ç”¨é»˜è®¤å€¼
try:
    MODEL_NAME
except NameError:
    MODEL_NAME = "gpt-4o"  # é»˜è®¤ä½¿ç”¨gpt-4oæ¨¡å‹

# åˆ›å»ºOpenAIå®¢æˆ·ç«¯å®ä¾‹
client = openai.OpenAI(api_key=API_KEY)

def get_completion(prompt: str):
    """
    è¾…åŠ©å‡½æ•°ï¼šå‘GPTå‘é€æç¤ºå¹¶è·å–å›å¤
    
    å‚æ•°:
        prompt (str): å‘é€ç»™GPTçš„æç¤ºæ–‡æœ¬
    
    è¿”å›:
        str: GPTçš„å›å¤æ–‡æœ¬
    """
    # æ„å»ºæ¶ˆæ¯åˆ—è¡¨
    messages = [{"role": "user", "content": prompt}]
    
    # è°ƒç”¨OpenAI API
    response = client.chat.completions.create(
        model=MODEL_NAME,           # ä½¿ç”¨çš„æ¨¡å‹åç§° (gpt-4o æˆ– deepseek-r1)
        messages=messages,          # æ¶ˆæ¯åˆ—è¡¨
        max_tokens=2000,           # æœ€å¤§ç”Ÿæˆtokenæ•°
        temperature=0.0            # æ¸©åº¦å‚æ•°ï¼Œ0è¡¨ç¤ºæ›´ç¡®å®šæ€§çš„è¾“å‡º
    )
    return response.choices[0].message.content
```

ç°åœ¨æˆ‘ä»¬å°†ä¸ºGPTç¼–å†™ä¸€ä¸ªç¤ºä¾‹æç¤ºï¼Œå¹¶é€šè¿‡è¿è¡Œæˆ‘ä»¬çš„`get_completion`è¾…åŠ©å‡½æ•°æ¥æ‰“å°GPTçš„è¾“å‡ºã€‚è¿è¡Œä¸‹é¢çš„å•å…ƒæ ¼å°†åœ¨å…¶ä¸‹æ–¹æ‰“å°å‡ºGPTçš„å“åº”ã€‚

è¯·éšæ„ä¿®æ”¹æç¤ºå­—ç¬¦ä¸²ï¼Œä»¥è·å¾—GPTçš„ä¸åŒå“åº”ã€‚

```python
# æç¤ºæ–‡æœ¬
prompt = "Hello, GPT!"

# è·å–GPTçš„å“åº”
print(get_completion(prompt))
```
